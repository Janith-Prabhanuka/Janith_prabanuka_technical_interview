### Question 6

```results_tbl``` is a Pandas table that shows the exam score of grade 10 students in 5 different classes.
Using pandas built-in functions, find out the average score value for each class.

````python
class_id = np.random.randint(5, size=20) + 1
student_id = np.array(range(len(class_id)))
exam_score = np.random.randint(100, size=len(class_id))

dct = {'Class ID':class_id, 'Student ID':student_id, 'Exam Score':exam_score}
results_tbl = pd.DataFrame(dct).set_index('Student ID')
results_tbl
````

````python
# Your Answer

results_tbl.groupby(['Class ID']).mean()
````

### Question 8

You are given a numpy multi-dimensional tensor; ```weights```. It is a floating point weight tensor and its values are in the range of <b>[-1000.0,1000.0]</b>.
You need to map (or transform) these weight values into <b>integer values</b> in the range <b>[0, 255]</b>.

Write a python code to perform this.
You can use any library or write it from scratch.

quantization_plot.png

````python
weights = np.random.uniform(low=-1000.0, high=1000.0, size=(16,3,7,7))
````

````python
def mapper(weight):
    weight/=1000 #Mapped between -1 to +1
    weight+=1 #Mapped between 0 to +2
    weight*=127.5
    return weight

print("Minimum Value : %.2f"%weights.min(),"Maximum Value : %.2f"%weights.max())
mapped_weights = mapper(weights)
print("Mapped Minimum Value : %.2f"%mapped_weights.min(),"Mapped Maximum Value : %.2f"%mapped_weights.max())
````
